```python
# load data
dtype = dict(names=('country','beer','spirit','wine','total'),
            formats=('S64','f','f','f','f'))
my_data_with_spec_colms = np.loadtxt("drinks.csv", delimiter=',', dtype=dtype, skiprows=1)
```

```python
my_data_with_spec_colms
```

```python
# Select data by country
my_data_with_spec_colms[np.where(my_data_with_spec_colms['country'] == 'Canada')]
```

```python
# Select data teetotal countries
my_data_with_spec_colms[np.where(my_data_with_spec_colms['total'] == 0)]
```

```python
# Who drinks the most?
my_data_with_spec_colms[np.where(my_data_with_spec_colms['wine'] == np.max(my_data_with_spec_colms['wine']))]
```

```python
# What is the average amount consumed?
np.mean(my_data_with_spec_colms['total'])
```

```python
# How much variation in the amount consumed?
np.std(my_data_with_spec_colms['total'])
```
